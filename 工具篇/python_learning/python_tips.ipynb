##  1. 随机数
python中产生随机数的主要有两个模块`random`和`np.random`

### 1.1 指定形状的随机数
- `numpy.random.rand(d0, d1, ..., dn)`  
Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).


````python
import numpy as np

print("shape with one dimension")
print(np.random.rand(1))

print("shape with two dimension")
print(np.random.rand(2, 3))

print("shape with three dimension")
x = np.random.rand(2, 3, 2)
print(x)
print(x.shape)
````

### 1.2 指定范围内指定大小的随机数
- `numpy.random.randint(low, high=None, size=None, dtype='l')`  
Return random integers from low (inclusive) to high (exclusive).
Return random integers from the “discrete uniform” distribution of the specified dtype in the “half-open” interval [low, high). If high is None (the default), then results are from [0, low).

````python
# 在范围[4, 10)之间产生20个随机数字
print(np.random.randint(low=4, high=10, size=20))

# 在范围[4, 10)之间产生个三维大小随机数字
print(np.random.randint(low=4, high=10, size=(2, 3, 2)))
````

### 1.3 从指定的范围内取出相应的数据
- `numpy.random.choice(a, size=None, replace=True, p=None)`
Generates a random sample from a given 1-D array

````python
# 从[1, 3, 5]中产生一个随机数字
print(np.random.choice([1, 3, 5], size=1))

# 从[1, 3, 5, 7, 9, 11]中产生一个二维矩阵数字（有放回）
print(np.random.choice([1, 3, 5, 7, 9, 11], size=(2, 3), replace=True))

# 从[1, 3, 5, 7]中产生一个二维矩阵数字（有放回）,每个元素取得的概率[0.05, 0.05, 0.1, 0.8]
print(np.random.choice([1, 3, 5, 7], size=(4, 5), replace=True, p=[0.05, 0.05, 0.1, 0.8]))
````

### 2.3 另外方式渲染中文

除了上面介绍到使用包`pltz`，可以下载相应的中文字体，然后绘图指定字体即可。发现在使用`pltz`包时，发现其功能和原生的`axes`有区别，仍存在一定问题，故而需求另外渲染中文的方式。

- 方式1 （测试通过）  
1.字体安装   
```
$ sudo apt-get install fonts-wqy-zenhei # 安装字体, 
$ cd /usr/share/fonts/truetype/wqy/ # cd到字体安装目录下
$ fc-list :lang=zh  # 查看新安装的字体的路径，在下面fname中添加
```
2.指定字体      
```
import matplotlib.pyplot as plt
import matplotlib.font_manager as mfm
zh_font = mfm.FontProperties(fname="/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc")
plt.xlabel(u'测试一下 ', fontsize=20, fontproperties=zhfont) # specify font 
```

- 方式2（不需要下载中文字体，推荐使用）   
在安装完包`pltz`时，会自带中文字体，使用方式`pltz.zh_font`，有关中文字体的大小等属性设置，可以`pltz.zh_font.<TAB>`查看

## 2.4 箭头和文字

````python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.linspace(0,1)
ax.plot(x, x, 'o')

ax.annotate('default line', xy=(0.15,0.1), xytext=(0.6,0.1),
            arrowprops={'arrowstyle': '-'}, va='center')
ax.annotate('dashed line', xy=(0.25,0.2), xytext=(0.6,0.2),
            arrowprops={'arrowstyle': '-', 'ls': 'dashed'}, va='center')
ax.annotate('default arrow', xy=(0.35,0.3), xytext=(0.6,0.3),
            arrowprops={'arrowstyle': '->'}, va='center')
ax.annotate('thick blue arrow', xy=(0.45,0.4), xytext=(0.6,0.4),
            arrowprops={'arrowstyle': '->', 'lw': 4, 'color': 'blue'},
            va='center')
ax.annotate('double-headed arrow', xy=(0.45,0.5), xytext=(0.01,0.5),
            arrowprops={'arrowstyle': '<->'}, va='center')
ax.annotate('arrow with closed head', xy=(0.55,0.6), xytext=(0.1,0.6),
            arrowprops={'arrowstyle': '-|>'}, va='center')
ax.annotate('a really thick red arrow\nwith not much space', xy=(0.65,0.7),
            xytext=(0.1,0.7), va='center', multialignment='right',
            arrowprops={'arrowstyle': '-|>', 'lw': 8, 'ec': 'r'})
ax.annotate('a really thick red arrow\nwith space between\nthe tail and the'
            'label', xy=(0.85,0.9), xytext=(0.1,0.9), va='center',
            multialignment='right',
            arrowprops={'arrowstyle': '-|>', 'lw': 8, 'ec': 'r', 'shrinkA': 10})

plt.show()
````

## 2.5 等高线图

````python
import matplotlib.pyplot as plt

f = lambda w1, w2: pow((w1-20), 2)+pow((w2-20), 2)

x=np.linspace(-30, 30, 100)
y=np.linspace(-30, 30, 100)
X,Y=np.meshgrid(x,y)

fig, ax = plt.subplots()
CS = ax.contour(X, Y, f(X, Y), 10)
ax.clabel(CS, inline=2, fontsize=8)
ax.set_title('contour plot')

ax.annotate('', xy=(x[-1],0), xytext=(x[0],0), va='center',
            multialignment='right',
            arrowprops={'arrowstyle': '-|>', 'lw': 4, 'ec': 'b', 'shrinkA': 1})

ax.annotate('', xy=(0, y[-1]), xytext=(0, y[0]), va='center',
            multialignment='right',
            arrowprops={'arrowstyle': '-|>', 'lw': 4, 'ec': 'b', 'shrinkA': 1})
````
